# Les thèmes

Dans la datavisualisation, le **style** est un aspect important qui facilite la lecture des graphique. **`seaborn`** à cet effet a intégré un ensemble de thème dans la librairie. Nous verrons dans la suite les thèmes disponibles et comment les utiliser. 

Les thèmes disponibles sont les suivants : **`darkgrid, whitegrid, dark, white, ticks`**.

Pour utiliser l'un de ces thèmes, il faut faire appel à la fonction **`sns.set_style("nom_theme")`** ou `nom_theme` correspond au nom du thème à utiliser.

Pour observer comment le thème affecte les graphique, construisons le nuage de points qui représente `sepal_length` en fonction de `sepal_width` en utilisant chacun de ces thèmes

```python
sns.set_style("darkgrid")
sns.scatterplot(data = iris_df, x = "sepal_width", y = "sepal_length")
```

```python
sns.set_style("whitegrid")
sns.scatterplot(x = iris_df.sepal_width, y = iris_df.sepal_length)
```

```python
sns.set_style("dark")
sns.scatterplot(x = iris_df.sepal_width, y = iris_df.sepal_length)
```

```python
sns.set_style("white")
sns.scatterplot(x = iris_df.sepal_width, y = iris_df.sepal_length)
```

```python
sns.set_style("ticks")
sns.scatterplot(x = iris_df.sepal_width, y = iris_df.sepal_length)
```

# Les graphiques

## Relation plot `sns.relplot`

Cette fonction permet de représenter la relation entre deux variables. 

Fonction complète ici : https://seaborn.pydata.org/generated/seaborn.relplot.html

```python
# définition du thème
sns.set_style("darkgrid")
```

```python
# construction du graphique
sns.relplot(data=iris_df, x="sepal_width", y="sepal_length")
```

```python
# Ajout de la différenciaiton des groupes 
sns.relplot(data=iris_df, x="sepal_width", y="sepal_length", hue="species")
```

```python
?sns.relplot
```

## Histogram `sns.histplot()`

Cette fonction représente la distribution des données (numériques). 

Pour tous les détails Cf. https://seaborn.pydata.org/generated/seaborn.distplot.html

```python
sns.histplot(data = iris_df, x="sepal_length")
```

```python
# Ajout de la courbe de densité
sns.histplot(data = iris_df, x="sepal_length", kde=True)
```

```python
?sns.histplot
```

## Bar Plot `sns.barplot()`

Réprésentre la relation entre deux variables, l'une étant catégorielle et l'autre numérique. 

```python
# Orientation verticale
sns.barplot(data=iris_df, x="species", y="sepal_length")
```

```python
# Orientation horizontale et sans intervalle de confiance
sns.barplot(data=iris_df, y="species", x="sepal_length", orient="h", ci=None)
```

```python
?sns.barplot
```

## Count Plot `sns.countplot()`

Représente pour une variable catégorielle la fréquence (nombre d'occurrences) de chaque modalité. On n'a pas besoin au préable de constuire une table qui pour chaque modalité donne la fréquence. Si on construit cette table, on peut directement utiliser un `barplot`

```python
sns.countplot(data=iris_df, x="species")
```

```python
?sns.countplot
```

```python
sns.pointplot(data=iris_df, x="species", y="sepal_length")
```

## Joint Plot `sns.jointplot()`

Ce graphique combine la relation entre deux variables quantitatives et la répartition de chaque variable sur le même graphique. 

```python
# nuage de points accompagné de l'histogramme
sns.jointplot(x='sepal_width', y='sepal_length', data=iris_df) 
```

```python
# Ajout de l'estimation de la droite de régression et de l'estimation de la courbe de densité
sns.jointplot(x='sepal_width', y='sepal_length', data=iris_df, kind='reg') 
```

## Regplot `sns.regplot()`

Ce graphique montre le nuage de point de deux variables et ajoute l'estimation de la droite de régression qui lie les deux variables.

```python
sns.regplot(x="sepal_width", y="sepal_length", data=iris_df)
```

```python
# suppression de l'intervalle de confiance
sns.regplot(x="sepal_width", y="sepal_length", data=iris_df, ci=None)
```

## Lm Plot `sns.lmplot()`

Tout comme `regplot`, la fonction `lmplot` estime et représente la droite de régression qui lie deux variables quantitatives. La différence avec `regplot` est qu'avec `lmplot` l'estimation peut se faire par catégorie.

```python
# Nuage de pionts accompagné de l'estimation de la droite de régression pour chaque catégories de fleurs
sns.lmplot(x="sepal_width", y="sepal_length", col="species", hue="species", data=iris_df, col_wrap=3, height=5)
```

## KDE Plot `sns.kdeplot()`

Cette fonction estime et représente la courbe de densité. 

```python
sns.kdeplot(data=iris_df, x="sepal_length", fill=True)
```

```python
# Ajout de la différenciation par catégorie
sns.kdeplot(data=iris_df, x="sepal_length", hue="species", fill=True)
```

## Box Plot `sns.boxplot()`

Cette fonction représente les boxplots. Ce graphique resume sur un schéma l'information sur les quartiles. Ce graphique est très souvent utilisé pour illustrer la présence des données atypiques. 

```python
sns.boxplot(x="species", y="sepal_length", data=iris_df)
```

## Heatmap `sns.heatmap()`

C'est un graphique à deux dimensions qui représente sur une échelle de couleur la valeur au croisement de la matrice. On peut l'utiliser par exemple pour illutrer la matrice de corrélation. 

```python
# Heatmap sur la matrice de corrélation
sns.heatmap(data=iris_df.drop(columns=["species"]).corr(), annot=True)
```

## Pair Plot `sns.pairplot()`
Ce graphique montre le nuage de points entre plusieurs variables accompagné de la l'estimation de la courbe de densité de chaque variable

```python
sns.pairplot(data=iris_df, diag_kind = "kde",kind = "scatter", palette = "husl")
```

```python
# ajout de la distinction par catégorie
sns.pairplot(data=iris_df, hue = 'species', diag_kind = "kde",kind = "scatter", palette = "husl")
```

# Paramètres du graphiques

## Labels des axes

- `set_title` pour les paramètres du titre (`fontsize`: pour la taille de police, `color` pour la couleur)
- `set_xlabel` pour le label de l'axe des X (`fontsize`: pour la taille de police, `color` pour la couleur)
- `set_xlabel` pour le label de l'axe des Y (`fontsize`: pour la taille de police, `color` pour la couleur)

```python
# boxplot
fig = sns.kdeplot(data=iris_df, x="sepal_length", hue="species", fill=True)

# Ajout du titre
fig.set_title("Sepal length distribution by Species category", fontsize=15, color='b')
 
# Ajout du label de l'axe Y
fig.set_ylabel( "Sepal Length")

# Ajout du label de l'axe X
fig.set_xlabel( "Species")
```

## Limites des axes

```python
fig = sns.kdeplot(data=iris_df, x="sepal_length", hue="species", fill=True)
fig.set_ylim(0, 0.5)
fig.set_xlim(0, 10)
```
